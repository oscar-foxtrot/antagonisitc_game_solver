<h3> -------------------------- <br> Oligopoly market matrix game</h3>

This is an example of a real-world problem.
The market of oligopolists is analyzed, represented by two competitors, K1 and K2. The controlled factors (strategies) of the competitors include production volumes, prices for the sold products, options for pricing or sales policies, advertising, etc. The effectiveness of decisions made by the competitors is assessed using the indicator of the Relative Market Share (RMS). If the market is completely divided among the oligopolists, then the percentage of the market that one competitor additionally gains is exactly the same percentage of the market that the other competitor loses. This satisfies the condition of antagonism.

```python
matrix = np.array(
                [[10, -2, -4],
                [6, 3, -5],
                [-8, 4, 2],
                [3, -9, 7]]
                )
solution = nash_equilibrium(matrix)
visualize(matrix, (solution[1], solution[2]))
```

<h3> -------------------------- <br>  Antagonistic Interests between Taxpayers and State Tax Authorities </h3>

Another real-world problem. In the context of antagonistic interests between taxpayers and state tax authorities, a conflict arises between two parties: A  the state tax inspection, and B  a taxpayer with a specific annual income, subjected to a tax of N. Each party employs its own set of strategies. The strategies of tax authorities are represented by columns and the strategies of taxpayers are represented by rows. It is assumed that the interests of taxpayers and tax authorities are directly opposing (antagonistic), particularly in situations involving negligent taxpayers. The gain for tax authorities, which corresponds to a loss for taxpayers, depends on the strategies employed by both tax authorities and taxpayers.

```python
matrix = np.array(
                [[8, 3, 4],
                [6, 5, 7],
                [4, 7, 6],
                [3, 10, 2]]
                )
solution = nash_equilibrium(matrix)
visualize(matrix, (solution[1], solution[2]))
```

<h3> -------------------------- <br>  Aaaand... The last example: Blotto game </h3>

The Blotto game is a strategic, two-player game that models competition or conflict over resource allocation. In this game, players distribute a fixed budget across multiple battlefields or resources, and the player who allocates their resources in the most strategically advantageous way wins.

```python
matrix = np.array(
                [[4, 0, 2, 1],
                [0, 4, 1, 2],
                [1, -1, 3, 0],
                [-1, 1, 0, 3],
                [-2, -2, 2, 2]]
                )
solution = nash_equilibrium(matrix)
visualize(matrix, (solution[1], solution[2]))
```
